```python
# read in tf record and display values
import tensorflow as tf
import numpy as np

dataset = tf.data.TFRecordDataset('./leap-complete-training-data-0004')


example = tf.train.Example()
example.ParseFromString(dataset.take(1).numpy())

# Iterate through features and print their types
for feature_name in example.features.feature:
    feature = example.features.feature[feature_name]
    dtype = tf.dtypes.as_dtype(feature.type)
    print(f"Feature '{feature_name}': {dtype}")
```
